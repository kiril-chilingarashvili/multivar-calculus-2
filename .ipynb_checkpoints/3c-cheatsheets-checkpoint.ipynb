## Flux - Lecture overview

## We've seen that we can integrate the tangential component of a **force field** $\vec{F}$ along a curve $C$, and this integral computes the **work** done by the force field to a point traveling along the curve $C$.

### In this lecture we explore what we get if we integrate the normal component of a vector field along a curve $C$. This situation arises when we consider the flow of particles through a boundary $C$ that move according to a **velocity field** $\vec{V}$. The line integral of the normal component of the velocity field along a curve is known as the **flux** through $C$.

### **Objectives**
### After this lecture and some practice, you will be able to
- ### Write the **flux** of a (velocity) vector field through a curve a as a **line integral** with respect to the outward normal of the curve.
- ### Use **Green's theorem in normal form** to compute flux via a double integral of the **divergence** of the vector field.
- ### Interpret flux as the amount of a quantity that passes through a boundary over unit time.
- ### Interpret divergence of a vector field as the amount of a quantity created or destroyed per unit time at any point.

## Flux defined as a line integral

### **Remark 2.1**
### For now, we define flux of 2D vector fields through curves in the plane. We will discuss flux of 3D vector fields through surfaces in the next course.
### Flux is computed via a line integral, a new type of line integral.

### **Definition 2.2**
### Given a curve $C$ in the plane, and a vector field $\vec{F}$, the **flux** of $\vec{F}$ across $C$ is
## $$ \int_C \vec{F}\cdot \hat{n} \mathop{ds} $$

### where $\hat{n}$ is the unit normal vector to the curve $C$ pointing $90^\circ$ clockwise from the unit tangent vector $\hat{T}$ to $C$.
![img](img/img-10-01.png)

### Note that there are two choices as to the direction that the unit normal vector can take. It can point to the left of the tangent vector or to the right of the tangent vector. Our convention assumes that the normal vector is always pointing to the right of the tangent vector to the curve.
![img](img/img-10-02.png)

### To compute the flux, we break the curve $C$ into small pieces of length $\Delta s$, then the flux is given by
## $$ \lim_{\Delta s\to 0} \sum \vec{F} \cdot \hat{n} \Delta s $$

## Comparison of flux and work

### Here we would like to compare the expression for flux to the expression we found for work. Note that they are both line integrals of a vector field.
## $$ \begin{array} {rcl} \text{Flux} & = & \displaystyle \int_C \vec{F}\cdot {\color{blue}{\hat{n}}} \mathop{ds} \\ \text{Work} & = & \displaystyle \int_C \vec{F}\cdot d\vec{r} = \int_C \vec{F} \cdot {\color{orange}{\hat{T}}} \mathop{ds} \end{array} $$

### Work is summing the ${\color{orange}{\text{tangential component}}}$ of the vector field $\vec{F}$ along the curve $C$. Roughly speaking, work measures how much your vector field is going along or against the curve.

### On the other hand, flux is summing the ${\color{blue}{\text{normal component}}}$ of the vector field $\vec{F}$ along the curve $C$. Roughly speaking it measures how much of the vector field is moving across the curve, where flux is considered positive if it moves to the right across the curve, and negative if it moves to the left across the curve.

## Physical interpretation of flux

### Let $\vec{F}$ be a velocity field for a fluid. Then at every point $(x,y)$ the vector $\vec{F}(x,y)$ points in the direction that the fluid is moving at that point, and the magnitude of that vector is the speed of fluid, with dimension Length per Time.

### Let $C$ be a curve.

### **Interpretation**:
### The flux of $\vec{F}$ across $C$ measures the net amount of fluid passing through $C$ (in the normal direction) per unit time.

### To see why this interpretation makes sense, we will analyze the amount of fluid that passes through a very small piece of curve, and see that we end up with the expression for flux we found on the previous page.

### Cut $C$ into small pieces of length $\Delta s$. If we cut our curve into small enough pieces, our curve will be essentially a straight line segment, and the velocity field $\vec{F}$ will be (essentially) constant along each piece.
![img](img/img-10-03.png)

### If we imagine each vector as a velocity vector of a fluid, then the amount of fluid passing through our small piece of curve is a parallelogram.
![img](img/img-10-04.png)

### The area of this parallelogram will correspond to the amount of fluid that passes through our curve in unit time. So let's try to understand how big this parallelogram is. The area of a parallelogram is the length of the parallelogram times the height of the parallelogram. In our case, the length is the arc length $\Delta s$, and the height is the component of $\vec{F}$ in the normal direction. That is, the area of this parallelogram is
## $$ \text{Area of parallelogram} = \text{base} \times \text{height} = \underbrace{\Delta s}_{\text{length of}\, C} \underbrace{(\vec{F} \cdot \hat{n})}_{\text{normal component of}\, \vec{F}} $$

### To compute the flux across a curve, we cut our curve into small pieces:
### 1. Compute the area of this parallelogram ($(\vec{F} \cdot \hat{n})\Delta s$) for each small piece,
### 2. Add them up.

## Positive and negative flux

### Note that implicit in this computation is the idea that when $\vec{F}\cdot \hat{n}$ points in the same direction as $\hat{n}$ the flux is positive. If $\vec{F}\cdot\hat{n}$ points in the opposite direction as $\hat{n}$, the flux is negative.

### Said another way, given a (tangent) direction along the curve $C$, the normal vector $\hat{n}$ points to the right of the of $C$.
- ### Fluid that moves from left to right across $C$ is counted as positive flux.
- ### Fluid that moves from right to left across $C$ is counted as negative flux.

### If we imagine a rotating flow in a circular tank, and $C$ is the diameter of the tank oriented from left to right, then the flux along the left half is negative and the right half is positive, yielding an overall net flux of $0$ because the two sides exactly cancel.

## Computing flux: geometric approach

### In the next two examples, we will explore the computation of flux from a geometric perspective. In both examples, we will consider the curve $C$ which is the circle of radius $a$ centered at the origin, and oriented counterclockwise.

### In this case, our curve is closed, so our line integral is over the closed curve. Note that the convention of using counterclockwise orientation gives us an outward pointing normal. It is convention to consider flux out of a region as positive when considering flux through closed curves.

### **Example 4.1**

### Let $C$ circle of radius $a$ centered at origin, oriented clockwise, and the vector field $\vec{F} = x\hat{i} + y\hat{j}$. Compute the flux of $\vec{F}$ out of $C$.
![img](img/img-10-08.png)

### Along $C$, $\vec{F}$ is parallel to (and pointing in the same direction as) the normal vector $\hat{n}$. Thus along $C$ we have
## $$ \vec{F}\cdot n = |\vec{F}|\underbrace{|\hat{n}|}_{=1} = |\vec{F}| $$
 
### Because the magnitude of $\vec{F}$ at a point is the distance of that point from the origin, along the circle of radius $a$, the magnitude is $a$:
## $$ |\vec{F}| = a $$
 
### Thus the flux out of $C$ is
## $$ \oint_C \vec{F}\cdot \hat{n} \mathop{ds} = \oint_C a \mathop{ds} = a\, \underbrace{\text{length of}\,C}_{2\pi a} = 2\pi a^2 $$

### Observe that this flux is positive, which is what we expect since we said that flow out of the circle would be positive.

### We will do one more geometric example with our other favorite vector field.

### **Example 4.2**

### Let $C$ circle of radius $a$ centered at origin, oriented clockwise, and the vector field $\vec{F} = -y\hat{i} + x\hat{j}$. Compute the flux of $\vec{F}$ out of $C$.
![img](img/img-10-09.png)

### Along $C$, $\vec{F}$ is tangent to the circle. Therefore it is perpendicular to the normal vector $\hat{n}$. Thus along $C$ we have
## $$ \vec{F}\cdot \hat{n} = 0 $$
 
### Thus the flux out of $C$ is $0$:
## $$ \oint_C \vec{F} \cdot \hat{n} \mathop{ds} = \oint_C 0\mathop{ds} = 0 $$

## Computing flux using components

### Recall from our computation of work, we wrote the vector $d\vec{r}$ symbolically in terms of components as
## $$ d\vec{r} = \hat{T}\mathop{ds} = \langle \mathop{dx}, \mathop{dy} \rangle $$

### Let's think about how we would write $\hat{n}\mathop{ds}$ similarly.
- ### Geometrically, $\hat{n}$ is $\hat{T}$ rotated by $90$ degrees clockwise.
- ### So it makes sense to think of $\hat{n}\mathop{ds}$ as $\hat{T} \mathop{ds}$ rotated by $90$ degrees clockwise.
![img](img/img-10-10.png)

### Thus we write symbolically
## $$ \boxed{\hat{n} \mathop{ds} = \langle \mathop{dy}, -\mathop{dx} \rangle } $$

### If $\vec{F} = \langle P, Q \rangle$, then
## $$ \int_C \vec{F}\cdot \hat{n}\mathop{ds} = \int_C \langle P, Q \rangle \cdot \langle \mathop{dy}, -\mathop{dx} \rangle = \int_C P\mathop{dy} - Q\mathop{dx} $$

### **Remark 5.1**
### In case this change of notation was confusing, we can write this flux integral in terms of our usual components $\vec{F} = \langle M, N \rangle$. The flux integral becomes
## $$ \int_C \vec{F}\cdot \hat{n}\mathop{ds} = \int_C \langle M, N \rangle \cdot \langle \mathop{dy}, -\mathop{dx} \rangle = \int_C M\mathop{dy} - N\mathop{dx} $$

## Green's theorem in normal form and divergence defined

### What if I need to compute flux, but I don't want to compute the line integral?

### **Upshot**: There is a version of Green's theorem for flux!

### **Green's theorem for flux, aka the 2D Divergence Theorem**
### **Theorem**
### If $C$ is a closed curve that encloses a region $R$ counterclockwise, and $\vec{F} = \langle P, Q \rangle$ is a vector field that is defined and differentiable on $R$, then the flux of $\vec{F}$ out of $R$ through $C$ is
## $$ \oint_C \vec{F}\cdot \hat{n}\mathop{ds} = \iint_R \mathrm{Div}\vec{F} \mathop{dA} $$
 
### The expression $\mathrm{Div}\vec{F}$ is a function of the vector field called the divergence of $\vec{F}$, which we will define now.

### **Definition 6.2**
### The divergence of a 2D vector field $\langle P, Q \rangle$ is
## $$ \mathrm{Div} \langle P, Q \rangle = P_x + Q_y $$

### **Spoiler alert: divergence in higher dimensions**
### The divergence of a 3D vector field $\langle P, Q, R \rangle$ is
## $$ \mathrm{Div}\langle P, Q, R \rangle = P_x + Q_y + R_x $$

### There are analogues of the divergence for higher dimensional vector fields, which you can look up independently.

### The divergence is sometimes written using the more compact notation
## $$ \mathrm{Div} \vec{F} = \nabla \cdot \vec{F} $$
 
### Note that $\nabla \cdot \vec{F}$ is a **scalar valued function**. In this case we interpret $\nabla$ as a differential operator
## $$ \nabla = \langle \frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z} \rangle $$
 
### Compare this to the gradient $\nabla f$, which is a vector field, but here $f$ is itself a scalar valued function, but $\nabla$ has the same interpretation.
## $$ \nabla f(x,y,z) = \langle \frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z} \rangle f(x,y,z) = \langle \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \rangle $$

### **Terminology**
### Note that Green's Theorem for **flux** is called
- ### **Green's theorem in normal form**.

### The **usual** Green's theorem for work is really
- ### Green's theorem in tangential form.

### (Green's theorem in normal form is also sometimes called the 2D divergence theorem.)

## Proof of Green's theorem in normal form

### The proof of Green's theorem in normal form is simpler than the proof in tangential form. Our proof will be notational, allowing us to rely on our earlier proof.

### **Proof**
### Let's start by writing out the result of the theorem in terms of components. Let $C$ be a closed curve oriented counterclockwise. Write $\vec{F} = \langle P, Q \rangle $, then we want to show that
## $$ \oint_C P\mathop{dy} - Q\mathop{dx} = \iint_R(P_x+Q_y)\mathop{dA} $$

### At its core, this is a relationship between a line integral and a double integral. Let's show that we can reduce it to the tangential form. We do this by ignoring the physical interpretation and focusing on the fact that the line integral is
## $$ \oint_C (\text{something})\mathop{dx} + (\text{something})\mathop{dy} $$
 
### Let's let
## $$ \begin{array} {rcl} M & = & -Q \\ N & = & P \end{array} $$

### Then
## $$ \oint_C \underbrace{(-Q)}_{M} \mathop{dx} + \underbrace{P}_{N} \mathop{dy} = \oint_C M\mathop{dx} + N\mathop{dy} $$
 
### By Green's theorem (in tangential form), this is equal to
## $$ \oint_C M\mathop{dx} + N\mathop{dy} = \iint (N_x - M_y)\mathop{dA} $$

### But we can think about what this is in terms of $P$ and $Q$. We have
## $$ \begin{array} {rcl} -M_y & = & Q_y \\ N_x & = & P_x \end{array} $$

### Therefore
## $$ \iint_R (\underbrace{N_x}_{P_x} \underbrace{-M_y}_{Q_y})\mathop{dA} = \iint_R(P_x + Q_y)\mathop{dA} $$
 
### And this is what we wanted to show.

### **Question**:
### **Is the way we compute these integrals independent of physical meaning?**
### Yes. At its core, we only have two types of integrals we can compute. We can compute line integrals, and we can compute double integrals. Once you have written a line integral as
## $$ \oint_C (\text{something})\mathop{dx} + (\text{something})\mathop{dy} $$
 
### you parametrize the curve, and use our standard line integral approach.

### Once you have written a double integral as
## $$ \iint_R (\text{something})\mathop{dA} $$

### you must figure out the bounds in terms of $x$ and $y$ and determine which to integrate first, or maybe use $r$ and $\theta$ in polar coordinates if it is easier. But at its core, it is the standard double integral that you are familiar with.

### The computations do not care about the physical interpretation as work or flux. The setup of the integrals themselves was where the physical interpretation was important.

## Applying Green's theorem to compute flux

### Here we look at an example that we've already seen. Earlier, we computed the flux of the vector field $\vec{F} = x\hat{i} + y\hat{j}$ through $C$ a circle of radius $a$ directly. Now we will compare this to what we can get if we use Green's theorem.

### **Example 8.1**

### Consider the velocity field $\vec{F} = x\hat{i} + y\hat{j}$ and $C$ the circle of radius $a$ oriented counterclockwise.
![img](img/img-10-11.png)

### To apply Green's theorem, we need to compute the divergence of $\vec{F}$.
## $$ \mathrm{Div} \vec{F} = \frac{\partial}{\partial x} x + \frac{\partial}{\partial y} y = 1 + 1 = 2 $$

### Green's theorem tells us that
## $$ \oint_C \vec{F} \cdot \hat{n}\mathop{ds} = \iint_R 2\mathop{dA} $$
 
### where $R$ is the disk of radius $a$.

### We know that $\displaystyle \iint_R \mathop{dA}$ is the area of the disk, which is $\pi a^2$. Therefore
## $$ \oint_C \vec{F}\cdot \hat{n}\mathop{ds} = \iint_R 2\mathop{dA} = 2 \underbrace{\pi a^2}_{\text{area of}\, R} $$
 
### This is the same answer that we saw before, which is good news.

### But we can do even better! Consider a curve $C$ which is a circle of radius $a$ that is no longer centered at the origin.
![img](img/img-10-12.png)

### This flux is harder to compute directly. Closest to the origin along the curve the flux is negative, but farther from the origin along the curve the flux is positive.
![img](img/img-10-13.png)

### Because the positive flux is larger in magnitude we suspect the net flux is positive, but this is harder to compute.

### However, if we use Green's theorem, this flux is much easier to compute. Notice that in our example above, we never used the fact that our circle of radius $a$ was centered at the origin. Because the divergence of $\vec{F}$ is the constant $2$ everywhere, it is true more generally that the flux out of any region is twice the area of that region.

### So the flux out of the circle of radius $a$ will be $2\pi a^2$ no matter where that circle is located.

## Physical interpretation of divergence

### As seen in the warm up problems, divergence is not sensitive to translational motion or rotational motions. What divergence is sensitive to is expanding or contracting motions.

### There are two possible physical interpretations, each makes sense in a different context.
### 1. Divergence measures how much the flow is **expanding** or **contracting** areas. This makes sense if you are considering the velocity of gases which can expand or contract to fill a volume.
### 2. Divergence measures the rate at which fluid is being pumped into the system. Sometimes called a **source rate** or **sink rate**. This makes sense if you are considering the velocity of water, which is incompressible.

### In the explanation above, the terms in orange indicate positive divergence, and the terms in blue indicate negative divergence. We will explore this idea and how divergence shows up when modeling conserved quantities will be explored in the recitation of the next lecture.

## Summary

### **Big Picture**
### Given a velocity field for a fluid, we can ask about the net rate per unit time at which fluid flows across a curve $C$. This net flow rate is known as **flux**. Our convention is that flux is positive when it flows across a parametrized curve from left to right. Flux is naturally described as a line integral across a curve.
## $$ \int_C \vec{F} \cdot \hat{n} \mathop{ds} $$

### Just as we were able to compute work via a line integral or an associated double integral via Green's theorem, we can apply **Green's theorem in normal form** to convert flux integrals, which are naturally written as line integrals, into double integrals of the **divergence** of the vector field over a region $R$ enclosed by the curve $C$ (oriented counterclockwise).
## $$ \int_C \vec{F} \cdot \hat{n} \mathop{ds} = \iint_R \mathrm{Div}\vec{F} \mathop{dA} $$

### The **divergence** of a vector field is a multivariable function that measures how much fluid is created or destroyed at any point.

### **Mechanics**
- ### To find a **unit normal vector** $\hat{n}$ to a curve $C$:
### 1. Take a parametrized $C$ as $(x(t), y(t))$.
### 2. The unit tangent vector is $\displaystyle \hat{T} = \frac{\langle x'(t), y'(t) \rangle}{\sqrt{x'(t)^2 + y'(t)^2}}$.
### 3. The unit normal $90$ degree clockwise from $\hat{T}$ is $\displaystyle \hat{n} = \frac{\langle y'(t), -x'(t) \rangle}{\sqrt{x'(t)^2 + y'(t)^2}}$.

- ### The divergence of a vector field $\vec{F} = \langle P, Q \rangle$ is
## $$ \mathrm{Div} \vec{F} = P_x + Q_y $$
 
### The divergence of a vector field is a multivariable (scalar-valued) function.

- ### Assuming $\vec{F}$ is defined and differentiable everywhere inside of a region $R$ enclosed by a counterclockwise oriented curve $C$, we can rewrite Green's theorem in terms of components as
## $$ \int_C -Q\mathop{dx} + P\mathop{dy} = \iint_R (P_x+Q_y)\mathop{dA} $$

### There are no new types of integrals introduced in this lecture. Flux can be computed as a line integral using standard line integral techniques, or as a double integral via Green's theorem.

### **Ask Yourself**
### **What is the net flux through a closed curve of a vector field with constant divergence?**
### By Green's theorem, for a vector field with constant divergence, the net flux is proportional to the area enclosed by the curve.

### **Is flux determined by the normal or tangential component of a vector field with respect to a curve?**
### Flux is measuring the fluid or particles that flow through a curve per unit length of curve, per unit time, and this depends on the normal component of the velocity field with respect to the curve.
