## Gradient fields

### Last time, we saw that the following were equivalent:
## $$ \begin{array} {rcl} \vec{F}\,\text{is a gradient field} & \iff & \displaystyle \int_C \vec{F} \cdot d\vec{r} \, \text{is path independent} \\ \, & = & \vec{F}\,\text{is conservative} \end{array} $$

### Let's recall what each of these terms means.
- ### A vector field $\vec{F}$ is a gradient field if and only if $\vec{F}(x, y) = \nabla f(x, y)$ for a function $f(x, y)$.
- ### The line integral for the work $\displaystyle \int_C\vec{F} \cdot d\vec{r}$ done by $\vec{F}$ along $C$ from point $P_0$ to $P_1$ is **path independent** if and only if the value of this end point depends only on the value of the potential function $f(x, y)$ at the end points
## $$ \int_C \vec{F}\cdot d\vec{r} = f(P_1) - f(P_0) $$
![img](img/img-8-01.png)

- ### The vector field $\vec{F}$ is **conservative** if and only if for any closed curve $C$,
## $$ \int_C \vec{F} \cdot d\vec{r} = 0 $$

### ***Goal***:
### Devise a test to determine when a vector field $\vec{F} = \langle M, N \rangle$ is a gradient field, and if so, find a function $f(x,y)$ so that $\nabla f = \vec{F}$.

### ***Objectives***
### After this lecture and some practice, you will be able to
- ### Define the **curl** of a vector field.
- ### Use the curl of a vector field in the plane to **determine when a vector field is a gradient field**.
- ### Recognize when we can and cannot use the curl to determine if a vector field is a gradient field.

## Exploration of gradient field

### **Goal**:
### Devise a test to determine when a vector field $\vec{F} = \langle M, N \rangle$ is a gradient field.

### **Exploration**:
### If $\vec{F} = \nabla f$, then
## $$ \begin{array} {rcl} M & = & f_x \\ N & = & f_y \end{array} $$

### When we learned partial derivatives, we learned that $f_{xy} = f_{yx}$. This tells us that if $\vec{F} = \langle M, N \rangle$ is a gradient field, then
## $$ \begin{array} {rcl} M_y & = & N_x \end{array} $$

### We will soon see that, conversely, if **unknown condition** and $M_y = N_x$, then $\vec{F}$ is a gradient field.

### For now, we will let our unknown condition be that $\vec{F}$ is defined everywhere and differentiable everywhere. We will find a more general condition at the end of this unit.

### For today:

### **Theorem**
### If $\vec{F} = \langle M, N \rangle$ is defined and differentiable everywhere **and** $M_y = N_x$, then $\vec{F}$ is a gradient field.

### ***Remark 3.2***
### You will see an example where the conclusion of the theorem fails in recitation. That is, you will see a vector field $\vec{F}$ where $M_y = N_x$, but there is a point where the vector field does not exist, and this point causes issues that cause it to fail to be a gradient field.

## Examples and practice

### **Example 4.1**

### Last time we decided that $\vec{F} = -y\hat{i} + x\hat{j}$ is not a gradient field because if you take any closed loop around the origin, the work done is positive.
![img](img/img-8-02.png)

### We check our property to verify that $M_y \neq N_x$.
## $$ \begin{array} {rcl} M_y & = & -1 \\ N_x & = & 1 \end{array} $$

### Therefore, since $M_y = -1 \neq 1 = N_x$, it follows that $\vec{F}$ is not a gradient field.

### **Question**: If we have $M_y = N_x$ on some region in the plane, is that enough to determine that our vector field is a gradient field?

### **Partial Answer**: For now, the general answer is no. (We will discuss this in detail in the last lecture in this unit.) The conclusion of our theorem fails to be true if the region where we are considering our vector field has a hole in it. (We will have to define this carefully later.) But, if for example, the region has no holes, called **simply connected**, then the conclusion of the theorem will hold and we will be able to describe our vector field as a gradient field on the simply connected region.

## Method 1: finding the potential function using line integrals

### Next logical question: Now that we have a vector field that is a gradient ($M_y = N_x$), how do we find the potential function?

### **Example 5.1**

### Let
## $$ \nabla f = \underbrace{(4x^2+8xy)}_{M}\hat{i} + \underbrace{(3y^2+4x^2)}_{N}\hat{j} $$
 
### How do we find $f$?

### There are two methods.

### **Method 1**
### **Compute line integrals.**
### Take some path that starts at the origin and go to a point $(x_1, y_1)$.
![img](img/img-8-03.png)
 
### By the fundamental theorem for line integrals, $\displaystyle \int_C \vec{F} \cdot d\vec{r} = f(x_1, y_1) - f(0,0)$. Therefore we can define
## $$ f(x_1, y_1) = \int_C \vec{F} \cdot d\vec{r} + \underbrace{f(0,0)}_{\text{constant}} $$
 
### We don't know the value of the constant, but in fact, we can define it to be any value that we want. When you find the antiderivative of a function, it is only defined up to adding a constant. We can compute the line integral using the definition of the line integral directly. (We learned how to compute these integrals in the last unit!)

### Note that we shouldn't take the path we described above. An easier path would be the straight line. But in fact, there is an even easier path, the path that goes from $(0,0)$ to $(x_1, 0)$ and then from $(x_1, 0)$ to $(x_1, y_1)$. Why is this easier? Remember that our line integral simplifies to $\displaystyle \int_C M \mathop{dx} + N\mathop{dy}$ and along each piece of this path, one half of the integrand is zero.

### **Example 5.2**

### Let $C$ be the path from $(0,0)$ that moves along the $x$-axis to $(x_1, 0)$, and then vertically to $(x_1, y_1)$. We let $C_1$ be the part of the path along the $x$-axis, and let $C_2$ be the vertical segment of the path.
![img](img/img-8-04.png)

### We want to compute $\displaystyle \int_C \vec{F} \cdot d\vec{r}$ for
## $$ \vec{F} = \underbrace{(4x^2+8xy)}_{M}\hat{i} + \underbrace{(3y^2+4x^2)}_{N}\hat{j} $$
 
### This becomes
## $$ \begin{array} {rcl} \displaystyle \int_C \vec{F} \cdot d\vec{r} & = & \displaystyle \int_C (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} \\ \, & = & \displaystyle \int_{C_1} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} \\ \, & & \displaystyle + \int_{C_2} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy}  \end{array} $$

### Along $C_1$, we have
- ### $x$ varies from $0$ to $x_1$
- ### $y=0$, and $dy = 0$

### Therefore the term $N$ in the integral is zero, and we are left with
## $$ \begin{array} {rcl} \displaystyle \int_{C_1} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} & = & \displaystyle \int_0^{x_1} 4x^2 \mathop{dx} \\ \, & = & \displaystyle \frac{4}{3}x_1^3 \end{array} $$

### Along $C_2$, we have
- ### $x=x_1$, and $dx = 0$
- ### $y$ goes from $0$ to $y_1$

### Therefore the term $M\mathop{dx}$ in the integral is zero, and we are left with
## $$ \begin{array} {rcl} \displaystyle \int_{C_2} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} & = & \displaystyle \int_0^{y_1} (3y^2+4x_1^2)\mathop{dy} \\ \, & = & y_1^3 + 4 x_1^2 y_1 \end{array} $$

### Therefore for any $(x_1, y_1)$, we have found that
## $$ f(x_1, y_1) = \underbrace{\frac{4}{3}x_1^3}_{\displaystyle \int_C M \mathop{dx}} + \underbrace{y_1^3+4x_1^2y_1}_{\displaystyle \int_C N \mathop{dy}} + f(0,0) $$
 
### Note that $f(0,0)$ can be any constant $c$. We can rewrite this in terms of $(x, y)$ in place of $(x_1, y_1)$, which were used to make the integration variables simpler and find a formula for any possible gradient:
## $$ f(x,y) = \frac{4}{3}x^3 + y^3 +4x^2y + c $$
 
### Note that if we are just looking for one potential function, we can omit the constant $c$. However, if we want to find every possible potential function, we want to keep the constant of integration, which gives us a way of describing every function that can be the potential function of the gradient field.

### **Remark 5.3**

### When applying this method, it is common to use a path that is piecewise, where each piece moves only in the $x$ or $y$ direction because it simplified our line integrals. However, one might have a function that is better described in polar coordinates. In such a case, a line that moves in the radial direction might be more natural. Similarly, for some reason, we might choose to move vertically in the $y$-direction first, then horizontally in the $x$-direction for a specific example where the line integrals simplify further. We will see examples using this approach later.
