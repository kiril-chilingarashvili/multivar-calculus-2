## Gradient fields

### Last time, we saw that the following were equivalent:
## $$ \begin{array} {rcl} \vec{F}\,\text{is a gradient field} & \iff & \displaystyle \int_C \vec{F} \cdot d\vec{r} \, \text{is path independent} \\ \, & = & \vec{F}\,\text{is conservative} \end{array} $$

### Let's recall what each of these terms means.
- ### A vector field $\vec{F}$ is a gradient field if and only if $\vec{F}(x, y) = \nabla f(x, y)$ for a function $f(x, y)$.
- ### The line integral for the work $\displaystyle \int_C\vec{F} \cdot d\vec{r}$ done by $\vec{F}$ along $C$ from point $P_0$ to $P_1$ is **path independent** if and only if the value of this end point depends only on the value of the potential function $f(x, y)$ at the end points
## $$ \int_C \vec{F}\cdot d\vec{r} = f(P_1) - f(P_0) $$
![img](img/img-8-01.png)

- ### The vector field $\vec{F}$ is **conservative** if and only if for any closed curve $C$,
## $$ \int_C \vec{F} \cdot d\vec{r} = 0 $$

### ***Goal***:
### Devise a test to determine when a vector field $\vec{F} = \langle M, N \rangle$ is a gradient field, and if so, find a function $f(x,y)$ so that $\nabla f = \vec{F}$.

### ***Objectives***
### After this lecture and some practice, you will be able to
- ### Define the **curl** of a vector field.
- ### Use the curl of a vector field in the plane to **determine when a vector field is a gradient field**.
- ### Recognize when we can and cannot use the curl to determine if a vector field is a gradient field.

## Exploration of gradient field

### **Goal**:
### Devise a test to determine when a vector field $\vec{F} = \langle M, N \rangle$ is a gradient field.

### **Exploration**:
### If $\vec{F} = \nabla f$, then
## $$ \begin{array} {rcl} M & = & f_x \\ N & = & f_y \end{array} $$

### When we learned partial derivatives, we learned that $f_{xy} = f_{yx}$. This tells us that if $\vec{F} = \langle M, N \rangle$ is a gradient field, then
## $$ \begin{array} {rcl} M_y & = & N_x \end{array} $$

### We will soon see that, conversely, if **unknown condition** and $M_y = N_x$, then $\vec{F}$ is a gradient field.

### For now, we will let our unknown condition be that $\vec{F}$ is defined everywhere and differentiable everywhere. We will find a more general condition at the end of this unit.

### For today:

### **Theorem**
### If $\vec{F} = \langle M, N \rangle$ is defined and differentiable everywhere **and** $M_y = N_x$, then $\vec{F}$ is a gradient field.

### ***Remark 3.2***
### You will see an example where the conclusion of the theorem fails in recitation. That is, you will see a vector field $\vec{F}$ where $M_y = N_x$, but there is a point where the vector field does not exist, and this point causes issues that cause it to fail to be a gradient field.

## Examples and practice

### **Example 4.1**

### Last time we decided that $\vec{F} = -y\hat{i} + x\hat{j}$ is not a gradient field because if you take any closed loop around the origin, the work done is positive.
![img](img/img-8-02.png)

### We check our property to verify that $M_y \neq N_x$.
## $$ \begin{array} {rcl} M_y & = & -1 \\ N_x & = & 1 \end{array} $$

### Therefore, since $M_y = -1 \neq 1 = N_x$, it follows that $\vec{F}$ is not a gradient field.

### **Question**: If we have $M_y = N_x$ on some region in the plane, is that enough to determine that our vector field is a gradient field?

### **Partial Answer**: For now, the general answer is no. (We will discuss this in detail in the last lecture in this unit.) The conclusion of our theorem fails to be true if the region where we are considering our vector field has a hole in it. (We will have to define this carefully later.) But, if for example, the region has no holes, called **simply connected**, then the conclusion of the theorem will hold and we will be able to describe our vector field as a gradient field on the simply connected region.

## Method 1: finding the potential function using line integrals

### Next logical question: Now that we have a vector field that is a gradient ($M_y = N_x$), how do we find the potential function?

### **Example 5.1**

### Let
## $$ \nabla f = \underbrace{(4x^2+8xy)}_{M}\hat{i} + \underbrace{(3y^2+4x^2)}_{N}\hat{j} $$
 
### How do we find $f$?

### There are two methods.

### **Method 1**
### **Compute line integrals.**
### Take some path that starts at the origin and go to a point $(x_1, y_1)$.
![img](img/img-8-03.png)
 
### By the fundamental theorem for line integrals, $\displaystyle \int_C \vec{F} \cdot d\vec{r} = f(x_1, y_1) - f(0,0)$. Therefore we can define
## $$ f(x_1, y_1) = \int_C \vec{F} \cdot d\vec{r} + \underbrace{f(0,0)}_{\text{constant}} $$
 
### We don't know the value of the constant, but in fact, we can define it to be any value that we want. When you find the antiderivative of a function, it is only defined up to adding a constant. We can compute the line integral using the definition of the line integral directly. (We learned how to compute these integrals in the last unit!)

### Note that we shouldn't take the path we described above. An easier path would be the straight line. But in fact, there is an even easier path, the path that goes from $(0,0)$ to $(x_1, 0)$ and then from $(x_1, 0)$ to $(x_1, y_1)$. Why is this easier? Remember that our line integral simplifies to $\displaystyle \int_C M \mathop{dx} + N\mathop{dy}$ and along each piece of this path, one half of the integrand is zero.

### **Example 5.2**

### Let $C$ be the path from $(0,0)$ that moves along the $x$-axis to $(x_1, 0)$, and then vertically to $(x_1, y_1)$. We let $C_1$ be the part of the path along the $x$-axis, and let $C_2$ be the vertical segment of the path.
![img](img/img-8-04.png)

### We want to compute $\displaystyle \int_C \vec{F} \cdot d\vec{r}$ for
## $$ \vec{F} = \underbrace{(4x^2+8xy)}_{M}\hat{i} + \underbrace{(3y^2+4x^2)}_{N}\hat{j} $$
 
### This becomes
## $$ \begin{array} {rcl} \displaystyle \int_C \vec{F} \cdot d\vec{r} & = & \displaystyle \int_C (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} \\ \, & = & \displaystyle \int_{C_1} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} \\ \, & & \displaystyle + \int_{C_2} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy}  \end{array} $$

### Along $C_1$, we have
- ### $x$ varies from $0$ to $x_1$
- ### $y=0$, and $dy = 0$

### Therefore the term $N$ in the integral is zero, and we are left with
## $$ \begin{array} {rcl} \displaystyle \int_{C_1} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} & = & \displaystyle \int_0^{x_1} 4x^2 \mathop{dx} \\ \, & = & \displaystyle \frac{4}{3}x_1^3 \end{array} $$

### Along $C_2$, we have
- ### $x=x_1$, and $dx = 0$
- ### $y$ goes from $0$ to $y_1$

### Therefore the term $M\mathop{dx}$ in the integral is zero, and we are left with
## $$ \begin{array} {rcl} \displaystyle \int_{C_2} (4x^2+8xy)\mathop{dx} + (3y^2+4x^2)\mathop{dy} & = & \displaystyle \int_0^{y_1} (3y^2+4x_1^2)\mathop{dy} \\ \, & = & y_1^3 + 4 x_1^2 y_1 \end{array} $$

### Therefore for any $(x_1, y_1)$, we have found that
## $$ f(x_1, y_1) = \underbrace{\frac{4}{3}x_1^3}_{\displaystyle \int_C M \mathop{dx}} + \underbrace{y_1^3+4x_1^2y_1}_{\displaystyle \int_C N \mathop{dy}} + f(0,0) $$
 
### Note that $f(0,0)$ can be any constant $c$. We can rewrite this in terms of $(x, y)$ in place of $(x_1, y_1)$, which were used to make the integration variables simpler and find a formula for any possible gradient:
## $$ f(x,y) = \frac{4}{3}x^3 + y^3 +4x^2y + c $$
 
### Note that if we are just looking for one potential function, we can omit the constant $c$. However, if we want to find every possible potential function, we want to keep the constant of integration, which gives us a way of describing every function that can be the potential function of the gradient field.

### **Remark 5.3**

### When applying this method, it is common to use a path that is piecewise, where each piece moves only in the $x$ or $y$ direction because it simplified our line integrals. However, one might have a function that is better described in polar coordinates. In such a case, a line that moves in the radial direction might be more natural. Similarly, for some reason, we might choose to move vertically in the $y$-direction first, then horizontally in the $x$-direction for a specific example where the line integrals simplify further. We will see examples using this approach later.

## Method 2: finding the potential using antiderivatives

### Our other method for finding potential functions involves using antiderivatives.

### Note that we know that we want to solve the system of partial differential equations
## $$ \begin{array} {lrcl} (1) & f_x & = & 4x^2+8xy \\ (2) & f_y & = & 3y^2 + 4x^2 \end{array} $$

### **Step 1: Take the antiderivative of $f_x$ with respect to** $x$
### The antiderivative of $(1)$ gives us an expression for $f(x,y)$
## $$ f(x,y) = \frac{4}{3} x^3 + 4x^2 y + g(y) $$

### In the case of antiderivatives of a multivariable function, our constant of integration is not a true constant. Instead, our constant of integration can be any function $g(y)$ that depends only on $y$.

### **Step 2: Take the partial derivative of $f$ with respect** $y$:
### We want to apply our condition from $(2)$. Since this condition involves a partial derivative of $f$ with respect to $y$, we take the partial $y$ derivative of what we found. This gives us
## $$ f_y = 4x^2 + g'(y) $$

### **Step 3: Match the two expressions for** $f_y$:
### We now have two expressions for $f_y$, so in order for them to be equal, we match them.
## $$ \begin{array} {rcl} 4x^2 + g'(y) & = & 3y^2 + 4x^2 \\ g'(y) & = & 3y^2 \\ \rightarrow g(y) & = & y^3 + c \end{array} $$

### Note that in this case, $g(y)$ can depend only on the variable $y$, so the constant of integration is a true constant, which we write here as $c$.

### **Step 4: Plug back into expression for** $f(x,y)$.
### Now we plug in our expression for $g(y)$ into our formula for $f(x,y)$ to obtain
## $$ f(x, y) = \frac{4}{3} x^3 + 4x^2 y + y^3 \,\,(+c) $$
 
### **Remark 6.1**

### You do not have to compute any line integrals with this method. However, you must be careful to adhere to the procedure.

### **Warning**: It is tempting to take an antiderivative of $f_x$ with respect to $x$ to find a formula $f(x,y)$ involving a function $g(y)$, then take an antiderivative of $f_y$ with respect to $y$ to find a formula $f(x,y)$ involving a function $h(x)$, and then attempt to match them. The issue with this approach is that if the functions involve sufficiently complicated trig functions, you may not immediately recognize that two function are equal up to a constant. For example $\tan^2(x)$ and $\sec^2(x)$ are equal up to a constant (because $1+\tan^2(x) = \sec^2(x)$).

### **You can start with $f_y$ instead of** $f_x$.
### 1. Take the antiderivative of $f_y$ to get $f(x,y) = \cdots + h(x)$.
### 2. Take the partial derivative of $f_y$ with respect to $x$.
### 3. Match the two expressions for $f_x$.
### 4. Plug into expression for $f$ found in step 1.

### **Pro tip**:
### Debugging your work 
### If in your process of matching the two expressions in step 3, you end up with something that depends on both $x$ and $y$, you made a mistake somewhere. At step 3 in your matching process, you will end up with a function of a single variable. So there must be an error somewhere.

### Note that the methods described here can be generalized to three dimensions (and higher). We will go through the three dimensional analogue of this theorem in the next course in this three part series.

## Introduction of curl

### The following are equivalent:
- ### $\vec{F} = \langle M, N \rangle$ is a gradient field
- ### $\vec{F}$ is conservative: that is for all closed curves $C$,
## $$ \oint_c \vec{F}\cdot d\vec{r} $$

### **New notation**:
### Note that we have introduced new notation. The notation $\displaystyle \oint_C$ is exactly the same as $\displaystyle \int_C$ as denoting a line integral. The $\displaystyle \oint_C$ includes the added reminder that the curve $C$ along which we are integrating is a closed curve.

### We have also seen that
## $$ \vec{F} = \langle M,N \rangle \, \text{is a gradient field} \quad \implies \quad N_x=M_y $$
 
### Similarly, if we have the additional condition that $\vec{F} = \langle M, N \rangle$ is defined and differentiable on the entire plane (or a simply connected region of the plane), then we get the opposite implication as well:
## $$ \text{Condition and}\, N_x = M_y \quad \implies \quad \vec{F} = \langle M, N \rangle\, \text{is a gradient field} $$

### In order to create a measurement of whether or not a vector field is or is not conservative, we introduce a new function.

### **Definition 8.1**

### The **curl** of a 2 dimensional vector field $\vec{F} = \langle M, N \rangle$ is a multivariable function defined by
## $$ \boxed{\mathrm{curl}(\vec{F}) = N_x - M_y} $$

### We can rewrite our condition on a vector field to determine if it is a gradient field in terms of the curl.
## $$ N_x = M_y \quad \text{becomes}\quad \mathrm{curl}(\vec{F}) = 0 $$
 
### We think of the curl of a vector field $\vec{F}$ as being a measurement of the failure of the vector field to be conservative.

### **Q and A**:
### **Is this curl the same curl as in physics?**
### Yes. This one is the same. Note however that this is the curl defined for a 2-dimensional vector field. There is a different formula for the curl of a 3-dimensional vector field (which we will see in the next course).

### **Should we always apply this test to check if a vector field is a gradient?**
### Yes. You may as well.

### Note that if $\mathrm(\vec{F})\neq 0$, you know automatically that your vector field is not a gradient, and is not conservative.
### However, if $\mathrm{curl}(\vec{F}) = 0$ then you still need to be careful about this additional hypothesis that the vector field be defined everywhere.

## Physical interpretation of curl

### **Curl of velocity fields**
### If $\vec{F}$ is a velocity field, then $\mathrm{curl}(\vec{F})$ measures the **rotation** component of the motion. If it is the velocity field of a fluid, for example, then the curl measures the vorticity, how much the motion twists around any given point.

### 1. Consider $\vec{F} = \langle a,b \rangle$ where $a$ and $b$ are constants.
![img](img/img-8-05.png)

### Then $\mathrm{curl}(\vec{F}) = 0$. Indeed the fluid is moving at one constant speed and direction, we do not think there should be any rotational component to this vector field.

### 2. Consider $\vec{F} = \langle x,y \rangle$.
![img](img/img-8-06.png)
 
### Then
## $$ \mathrm{curl}(\vec{F}) = \frac{\partial}{\partial x} y - \frac{\partial}{\partial y}x = 0 $$
 
### Note that in this case as well, the motion is radially outward from the origin, and we do not expect any rotational motion.

### 3. If instead we consider our favorite rotational vector field, $\vec{F} = \langle -y, x \rangle$,
![img](img/img-8-07.png)
 
### then the curl is
## $$ \mathrm{curl}(\vec{F}) = \frac{\partial}{\partial x} x - \frac{\partial}{\partial y}(-y) = 1 + 1 = 2  $$
 
### In fact we know that this vector field is moving with unit angular velocity. So the curl is measuring twice the angular velocity at any point.

### **Remark 9.1**
### For a class of vector fields called conformal vector fields, the motion can be decomposed into translation, dilation, and rotation only. In general, there may also be shear effects. Even more generally, there is the Helmholtz decomposition, which states that all vector fields can be decomposed into a sum
## $$ \vec{F} = \vec{G} + \vec{H} $$
 
### such that $\vec{G}$ is a gradient field, and $\vec{H}$ is **divergence free**. At this point, we have not defined the notion of divergence. We will define it in this unit, and can explore this idea more fully then. The idea is that $\vec{G}$ has no curl. (That is what we've discussed in this page, a vector field that is a gradient has zero curl.) The $\vec{G}$ component of the vector field is the part of the vector field that is free of rotation. The vector field $\vec{H}$ contains all of the rotational components, and has an additional property that it doesn't contain any dilation effects (i.e. it is divergence free).

### **Examples 9.2**

### If you are looking at a vector field that represents wind velocity, regions of high curl will be tornados and hurricanes. The sign of the curl, whether it is positive or negative, tells you the direction of rotation. Clockwise rotation is negative, and counterclockwise rotation is positive.

### **Curl of a velocity field conclusion**:
### Curl measures (twice) the angular velocity of the rotational component of a velocity field.

### **Curl of a force field**
### **Curl of a force field conclusion**:
### The curl of a force field measures the **torque** exerted on a test object that you put at any point in the field.

### **Physics analogy**
### There is an analogy between force and torque, between velocity and angular velocity, and between mass and moment of inertia.

### Generally in physics, we think of
## $$ \text{Force} = \frac{d}{dt}(\text{mass}\cdot \text{velocity}) $$
 
### (Note that often the mass is independent of time, and this turns into our general formula $\displaystyle F = m\frac{d}{dt}v = ma$.)

### Similarly, we have an analogous expression for torque:
## $$ \text{Torque} = \frac{d}{dt}(\text{moment of inertia})\cdot(\text{angular velocity}) $$
 
### (Note that often the moment of inertia is independent of time, and this turns into our general formula for Torque.)

### We've seen that curl measures the angular velocity of a velocity field. This analogy allows us to gain some intuition for the idea that curl might measure the torque of a force field.

## Summary

### **Big Picture**
### Given a vector field $\vec{F}$, if it satisfies some minimal hypotheses on a region of the plane, then the following are equivalent:
### 1. $\vec{F}$ is conservative ($\displaystyle \oint_C \vec{F}\cdot d\vec{r} = 0$ for any closed curve),
### 2. $\vec{F}$ is path independent,
### 3. $\vec{F}$ is a gradient field ($\vec{F} = \nabla f(x,y)$),
### 4. $\mathrm{curl} (\vec{F}) = 0$ **and** $\vec{F}$ is defined and differentiable everywhere.

### Physically, we care about conservative vector fields because many physical quantities are conserved, that is they cannot be created or destroyed: energy, momentum, mass. Thus in the development of physical models, we would expect our models to retain these modes of conservation, giving us fields whose curl is zero.

### The curl of a vector field is itself a multivariable function, which measures how far a vector field deviates from being conservative.

- ### If the vector field is a velocity field, the curl measures the angular velocity of the motion.
- ### If the vector field is a force field, the curl measures the torque per unit moment of inertia on a test object placed in that force field. (That is it measures how the angular velocity is increasing or decreasing.)

### Once we know that our vector field is a gradient field, we can find the potential function.

### **Mechanics**
### Given a vector field $\vec{F} = \langle M,N \rangle$, to test that this vector field is a gradient field (i.e. is conservative), it is enough to check the following:
### 1. The vector field is defined (and differentiable) everywhere (or on a simply connected region of the plane).
### 2. $\mathrm{curl}(\vec{F}) = N_x - M_y = 0$

### To find the potential function associated with a gradient field, we discussed two methods.
### 1. Define $\displaystyle f(x,y) = \int_C \vec{F} \cdot d\vec{r} + f(P_0)$ for any curve $C$ from $P_0$ to $(x,y)$. Usually we use a curve that makes the integrals easy to compute.
### 2. Anti-differentiate using the fact that $\vec{F} = \langle f_x, f_y \rangle$.
###    - Anti-differentiate $f_x$ with respect to $x$ to find a function $f(x,y) + g(y)$.
###    - Take the partial derivative of the first step with respect to $y$ and compare with your expression for $f_y$.

### **Ask Yourself**
### **Can a vector field have zero curl, but not be a gradient field?**
### Yes!

### **Can a vector field be conservative, but not be a gradient field?***
### No! A vector field being conservative is logically equivalent to being a gradient field.
