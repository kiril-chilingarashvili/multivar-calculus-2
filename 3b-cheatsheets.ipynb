## Green's Theorem - Lecture Overview

### So far, we've learned two types of integrals.
- ### Double integrals add up quantities over a 2D region.
- ### Line integrals add up quantities along a 1D curve.

### In this lecture we will introduce Green's theorem, which relates these two types of integrals in the case that we are looking at the line integral of a vector quantity along a closed curve, and the vector field is well defined on the entire enclosed region.

## Statement of Green's theorem

### **Review**
### Last time, given a vector field $\vec{F} = \langle M, N \rangle$, we defined the **curl** of the vector field to be
## $$ \mathrm{curl}(\vec{F}) = N_x-M_y $$

### We said that the curl measures how far the vector field is from being **conservative**. The goal of our last lecture was to determine the necessary conditions
- ### $\mathrm{curl}(\vec{F}) = 0$, and
- ### $\vec{F}$ is defined and differentiable everywhere
### so that we know that the vector field $\vec{F}$ is conservative.

### Once we know that a vector field is conservative, we know that the line integral around any closed curve is zero:
## $$ \oint_C \vec{F}\cdot d\vec{r} = 0 $$
 
### **Goal**
### Suppose we have a vector field $\vec{F}$ that is not conservative, that is $\mathrm{curl}(\vec{F}) \neq 0$. Suppose also that we have a closed curve $C$. Then in order to compute
## $$ \oint_C \vec{F}\cdot d\vec{r} $$
 
### we have two options.
### 1. First option is direct computation of the line integral.
### 2. The second option is the topic for today. We can apply Green's theorem, which involves computing a double integral instead.

### **Green's Theorem**
### **Theorem**
### If
- ### $C$ is a closed curve, enclosing a region $R$ counterclockwise,
![img](img/img-9-01.png)
 
### and

- ### $\vec{F}$ is a vector field that is defined and differentiable on $R$,

### then
## $$ \boxed{\oint_C \vec{F}\cdot d\vec{r} = \iint_R \mathrm{curl}(\vec{F})\mathop{dA}} $$

### Writing $\vec{F} = \langle M, N \rangle$, we can write this in terms of the components as
## $$ \boxed{\oint_C M\mathop{dx} + N\mathop{dy} = \iint_R(N_x-M_y)\mathop{dA}} $$

### This statement seems very strange! How can we relate a line integral to a double integral?

### **Remark 2.2**
### Why does the theorem specify that $C$ move counterclockwise around $R$? This is a matter of convention. The reason is because of our convention for curl as $N_x-M_y$ and not the other way around. The two conventions match with each other. You can imagine an alien might have the opposite convention, and their theorem would involve clockwise curves and integrals of $M_y-N_x$.

## Applying Green's theorem to compute work

### **Warning**!
### Green's theorem only holds for **closed** curves $C$.

### **Example 3.1**
### Let $C$ be the circle of radius $1$ centered at the point $(2.0)$, traversed counterclockwise.
![img](img/img-9-05.png)

### Compute
## $$ \oint_C \underbrace{ye^{-x}}_{M} \mathop{dx} +\underbrace{\frac{1}{2} x^2-e^{-x}}_{N} \mathop{dy}  $$
 
### **Solution: direct approach**
### Parametrize $C$:
## $$ \begin{array} {rcl} x = \cos(\theta)+2, & \, & \mathop{dx} = -\sin(\theta)\mathop{d\theta} \\ y = \sin(\theta), & \, & \mathop{dy} = \cos(\theta)\mathop{d\theta} \end{array} $$

### Substitute into the expression for the line integral and integrate as $\theta$ ranges from $0$ to $2\pi$. You can see already that this will be very complicated as you will end up trying to integrate an expression involving terms of the type $e^{\cos(\theta)+2}$. We will not try this!

### **Solution: Apply Green's Theorem**

### Instead we compute the double integral
## $$ \iint_R \mathrm{curl}(\vec{F})\mathop{dA} = \iint_R (N_x-M_y)\mathop{dA} $$
 
### Note that we need to compute the partial derivatives:
## $$ \begin{array} {rcl} N_x & = & x+e^{-x} \\ M_y & = & e^{-x} \end{array} $$

### Then we can write this as
## $$ \begin{array} {rcl} \displaystyle \iint_R (N_x-M_y)\mathop{dA} & = & \displaystyle \iint_R ((x+e^{-x})-(e^{-x})) \mathop{dA} \\ \, & = & \displaystyle \iint_R x\mathop{dA} \end{array} $$

### At this point, we could try to solve this by finding bounds for the region $R$. But instead, we recall that the center of mass of a region $R$ with mass density $\delta$ is the point $(\overline{x}, \overline{y})$ is defined such that
## $$ \overline{x} = \frac{\displaystyle \iint_R x\delta \mathop{dA}}{\text{Mass of}\, R} $$

### In the case that the density is uniform, $\delta = 1$, this reduces to an area computation
## $$ \overline{x} = \frac{\displaystyle \iint_R x \mathop{dA}}{\text{Area of}\, R} $$

### Therefore we can rearrange this expression to see that the integral we want to compute is
## $$ \iint_R x\mathop{dA} = \overline{x} \cdot \text{Area of }\, R = 2\pi $$
 
### Note that $\overline{x} = 2$ because of symmetry, and the area of a unit circle is $\pi$.

## Understanding Green's Theorem

### **Review**
### In the last unit, we started proving that the following were equivalent:
### 1. $\vec{F}$ is conservative.
### 2. $\displaystyle \int_C \vec{F}\cdot d\vec{r}$ is path-independent.
### 3. $\vec{F}$ is a gradient field $\vec{F} = \nabla f$.
### 4. $M\mathop{dx} + N\mathop{dy}$ is an exact differential.

### A differential $M\mathop{dx} + N\mathop{dy}$ is exact if and only if $M_y = N_x$, which holds if and only if $\mathrm{curl}(\vec{F}) = 0$. Thus we can rewrite the last statement as $N_x-M_y = 0$.

### In the last unit, we showed $(1)\iff (2)$, and $(3)\implies(2)$, and sketched how you would show $(2)\implies (3)$. In the previous lecture, we saw that $(3)\implies(4)$. The results of this page allow you to apply Green's theorem to show that $(4)\implies(1)$. Because $(1)$ is equivalent to $(3)$, this finishes the proof.

### Now that we've seen an example of how to apply Green's theorem, let us apply this theorem to prove that when $\mathrm{curl}(\vec{F}) = 0$, the vector field $\vec{F}$ is conservative.

### **Consequence of Green's theorem**: 
### If $\vec{F}$ is defined everywhere in the plane, and $\mathrm{curl}(\vec{F}) = 0$ everywhere in the plane, then $\vec{F}$ is conservative.

### **Proof sketch**:
### Take any (simple) closed curve $C$ in the plane.

### If $C$ is oriented clockwise, change the orientation to be counterclockwise. Since $\vec{F}$ is defined and differentiable everywhere, in particular we have that $\mathrm{curl}(\vec{F})=0$ everywhere on the region $R$ enclosed by $C$. We can apply Green's theorem, which tells us that
## $$ \oint_C \vec{F}\cdot d\vec{r} = \iint_R \mathrm{curl}(\vec{F})\mathop{dA} = \iint_R 0\mathop{dA} = 0 $$
 
### Therefore, since this holds for any curve $C$, we see that $\vec{F}$ is conservative.

### This proves that if $\mathrm{curl}(\vec{F}) = 0$ everywhere, then $\displaystyle \oint_C \vec{F}\cdot d\vec{r} = 0$.

### **What can go wrong?**
### Let's see what can go wrong if the hypotheses of the theorem do not hold!

### In the last recitation, we did a problem looking at the vector field
## $$ \vec{F} = \frac{-y\hat{i}+x\hat{j}}{x^2+y^2} $$
 
### You computed $\mathrm{curl}(\vec{F})$ and saw that it was zero everywhere. However, for $C$ the unit circle, you computed
## $$ \oint_C \frac{-y\mathop{dx}+x\mathop{dy}}{x^2+y^2} = 2\pi $$
 
### This function failed the hypothesis that $\vec{F}$ is defined and differentiable everywhere on $R$ because it is undefined at the origin $(0,0)$. Somehow, because the vector field is not defined on this one point inside the region enclosed by the curve $C$, we cannot apply Green's theorem.

### **Remark 4.1**
### The full proof requires the Jordan Curve Theorem and other details omitted here.

### For example, this theorem does not apply directly to curves that cross themselves. For example, if you wish to apply Green's theorem to a curve such as a figure 8 curve below,
![img](img/img-9-06.png)

### you must split the curve into two pieces and apply Green's theorem to each piece independently.

### Notice that if you split it up, one piece is traversed counterclockwise, but the other is traversed clockwise, which is what poses an issue with applying Green's theorem directly, without splitting it up.
![img](img/img-9-07.png)

## Proof of Green's theorem

### **Part 1: Reduction of complexity**
### We want to prove that
## $$ \oint_C M\mathop{dx} + N\mathop{dy} = \iint_R(N_x-M_y)\mathop{dA} $$
 
### To prove this statement, we will start by making a series of simplifications that will allow us to prove a simpler statement.

### **Observation 1**:
### We can reduce to the simpler case that $N=0$ and prove that
## $$ \oint_C M\mathop{dx} = \iint_R -M_y \mathop{dA} $$
 
### Why is this enough? We will be able to modify the argument to prove the special case $M=0$ in an analogous way, showing
## $$ \oint_C N\mathop{dy} = \iint_R N_x\mathop{dA} $$
 
### By adding the two results together, we obtain Green's theorem.

### **Observation 2**: We can decompose $R$ into simpler regions.
![img](img/img-9-08.png)

### If a curve $C$ encloses a region $R$, we can break the region $R$ into pieces $R_1$ and $R_2$ bounded by curves $C_1$ and $C_2$ respectively.
![img](img/img-9-09.png)

### If we can prove Green's theorem for $C_1$ and $R_1$ and $C_2$ and $R_2$, then we can prove the theorem is true for $C$ and $R$. That is if we can show that both
## $$ \oint_{C_1} M\mathop{dx} = \iint_{R_1} -M_y \mathop{dA} $$
 
### and
## $$ \oint_{C_2} M\mathop{dx} = \iint_{R_2} -M_y \mathop{dA} $$
 
### then Green's theorem follows because
## $$ \oint_C M\mathop{dx} = \oint_{C_1} M\mathop{dx} + \oint_{C_2} M\mathop{dx} = \iint_{R_1} -M_y \mathop{dA} + \iint_{R_2} -M_y \mathop{dA} = \iint_R -M_y \mathop{dA} $$


### Observe that the first equality holds because both $C_1$ and $C_2$ contain an extra curve that is the boundary between $R_1$ and $R_2$. Notice that we go along that curve twice, but with opposite orientations, so it cancels.

### **Observation 3**: If we cut up our region $R$ into "vertically simple" regions, it is simple to write the bounds for the double integral.
![img](img/img-9-10.png)

### A vertically simple region $R$ is one for which there are numbers $a$ and $b$ and functions $f_z(x)$ and $f_2(x)$ such that the region is described by $a<x<b$, and $f_1(x)<y<f_2(x)$.
![img](img/img-9-11.png)

### On such a vertically simple region, the double integral can be expressed as
## $$ \iint_R -M_y \mathop{dA} = \int_a^b \int_{f_1(x)}^{f_2(x)}-M_y \mathop{dy}\mathop{dx} $$

### **Proof Part 2: the main argument**
### We now prove that
## $$ \oint_C M\mathop{dx} = \iint_R -M_y \mathop{dA} $$

### in the case that $R$ is vertically simple, and $C$ is the boundary of $R$ going counterclockwise.

### Our vertically simple region looks like a region that is bounded by a value $x=a$ on the left and $x=b$ on the right. It is bounded by a curve $y=f_2(x)$ from above, and $y=f_1(x)$ from below.
![img](img/img-9-12.png) 

### To prove the equality, we will compute both sides, and see that they are equal.

### **The line integral**:

### Write the curve $C$ as a sum of four curves $C_1$, $C_2$, $C_3$ and $C_4$ such that $C_1$ is the bottom curve where $y=f_1(x)$ and going counterclockwise around the region $R$.
![img](img/img-9-13.png)

- ### Along $C_1$, we know that $y=f_1(x)$, and $x$ goes from $a$ to $b$, so we can write the line integral in terms of $x$:
## $$ \int_{C_1}M(x,y) \mathop{dx} = \int_a^b M(x, f_1(x)) \mathop{dx} $$
 
### This is as far as we can simplify things since we do not have an expression for $f_1(x)$.
- ### Along $C_2$, $\mathop{dx} = 0$, so
## $$ \int_{C_2} M\mathop{dx} = 0 $$
 
- ### Similarly, along $C_4$, $\mathop{dx} = 0$, so
## $$ \int_{C_4} M\mathop{dx} = 0 $$
 
- ### Along $C_3$, we know that $y=f_2(x)$, and $x$ goes from $b$ to $a$, so we can write the line integral in terms of $x$:
## $$ \int_{C_3} M(x,y) \mathop{dx} = \int_b^a M(x, f_2(x)) \mathop{dx} = -\int_a^b M(x, f_2(x)) \mathop{dx} $$
 
### Adding these pieces together, we get
## $$ \int_C M\mathop{dx} = \int_a^b M(x, f_1(x)) \mathop{dx} - \int_a^b M(x, f_2(x)) \mathop{dx} $$

### **The double integral**:
### Using the fact that our region is vertically simple, we set up the double integral using $dA = \mathop{dy}\mathop{dx}$ to get
## $$ \iint_R M_y \mathop{dA} = \int_a^b \int_{f_1(x)}^{f_2(x)}\frac{\partial M}{\partial y} \mathop{dy}\mathop{dx} $$

### By the fundamental theorem (single variable version), the inner integral gives us
## $$ \int_{f_1(x)}^{f_2(x)} \frac{\partial M}{\partial y} \mathop{dy} = M(x, f_2(x)) - M(x, f_1(x)) $$

### Pulling out a negative sign, we can write the full double integral as
## $$ \int_a^b \int_{f_1(x)}^{f_2(x)} - \frac{\partial M}{\partial y} \mathop{dy}\mathop{dx} = \int_a^b (M(x, f_2(x)) - M(x, f_1(x))) \mathop{dx} $$

### Thus the left hand side (the line integral) and the right hand side (the double integral) are equal to the same expression on a vertically simple region. Therefore they are equal for all vertically simple regions.

### This is the end of the proof, because we can then remove the vertically simple requirement by observations 2 and 3 above. And then remove the restriction on the type of vector field by showing the analogous statement holds for $N$ on horizontally simple regions.

## Summary

### **Big Picture**
### The necessary ingredients are
- ### a closed curve $C$ that bounds a region $R$ in the counterclockwise direction, and
- ### a vector field $\vec{F}$ defined and differentiable everywhere on $R$.

### Then Green's theorem says
## $$ \oint_C \vec{F}\cdot d\vec{r} $$

### This holds for any vector field satisfying the necessary hypotheses!

### This theorem is also the missing piece of theory needed to show that if the curl of a vector field is zero, then the vector field is conservative.

### The big picture is that somehow the curl, which measures the failure of a vector field to be conservative, can also measure the work done along a **closed**, **counterclockwise** curve. This powerful result reduces computations we developed in Unit 2 to computations we learned in Unit 1.

### **Mechanics**

### If we have closed curve $C$ that bounds a region $R$ in the counterclockwise direction, and a vector field $\vec{F} = \langle M, N \rangle$ defined and differentiable everywhere on $R$, then written in terms of coordinates, Green's theorem says
## $$ \oint_C M\mathop{dx} + N\mathop{dy} = \iint_R (N_x - M_y) \mathop{dA} $$
 
### **Ask Yourself**
### **Can Green's theorem be applied on regions where the vector field is not defined?**
### No! This is a necessary hypothesis. If the vector field (or its curl) are not defined on the region $R$ bounded by a curve $C$, then we cannot apply the theorem.

### **Can you apply Green's theorem on clockwise curves?**
### Sure! Remember that if $C$ is clockwise, then $-C$ is counterclockwise. Apply Green's theorem to find that
## $$ \oint_{-C} \vec{F}\cdot d\vec{r} = \iint_R \mathrm{curl}(\vec{F})\mathop{dA} $$
 
### Then use the fact that
## $$ \oint_{-C} \vec{F}\cdot d\vec{r} = -\oint_{C} \vec{F}\cdot d\vec{r} $$
 
### to obtain the final result.


### **Is it possible that you can apply Green's theorem on some closed curves, but not all closed curves?**
### Yes! If you have a vector field that is not defined or differentiable at a few points on the plane, that is OK! You won't be able to apply Green's theorem on any closed curves that contain points where the vector field or its curl are undefined, but for all other regions, you can apply the theorem!
